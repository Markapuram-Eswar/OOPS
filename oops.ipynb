```python
class Bank:
    def __init__(self, name, age, branch, gender, balance, account_number, pin):
        self.name = name
        self.age = age
        self.branch = branch
        self.gender = gender
        self.balance = balance
        self.account_number = account_number
        self.pin = pin

    def deposit(self):
        amount = int(input("Enter the amount to deposit: "))
        pin = int(input("Enter the pin: "))
        if pin == self.pin:
            self.balance += amount
            print(f"Deposited {amount} into {self.name}'s account")
        else:
            print("Invalid pin")

    def withdraw(self):
        amount = int(input("Enter the amount to withdraw: "))
        pin = int(input("Enter the pin: "))
        if pin == self.pin:
            if self.balance >= amount:
                self.balance -= amount
                print(f"Withdrew {amount} from {self.name}'s account")
            else:
                print("Insufficient balance")
        else:
            print("Invalid pin")

    def display(self):
        print(f"Name: {self.name}, Age: {self.age}, Branch: {self.branch}, "
              f"Gender: {self.gender}, Balance: {self.balance}, Account: {self.account_number}")

    def check_balance(self):
        pin = int(input("Enter the pin: "))
        if pin == self.pin:
            print(f"Current balance: {self.balance}")
        else:
            print("Invalid pin")

    def transfer(self, accounts):  # Pass all accounts as a parameter
        recipient_acc_num = input("Enter the account number of the recipient: ")
        amount = int(input("Enter the amount to transfer: "))
        pin = int(input("Enter the pin: "))

        if pin != self.pin:
            print("Invalid pin")
            return

        if recipient_acc_num == self.account_number:
            print("You cannot transfer money to your own account")
            return

        # Find the recipient account
        recipient = None
        for acc in accounts:
            if acc.account_number == recipient_acc_num:
                recipient = acc
                break

        if recipient is None:
            print("Recipient account not found")
            return

        if self.balance < amount:
            print("Insufficient balance")
            return

        # Perform transfer
        self.balance -= amount
        recipient.balance += amount
        print(f"Transferred {amount} to {recipient.name}'s account")

    def change_pin(self):
        old_pin = int(input("Enter the old pin: "))
        if old_pin == self.pin:
            new_pin = int(input("Enter the new pin: "))
            self.pin = new_pin
            print("Pin changed successfully")
        else:
            print("Invalid old pin")


# Example usage:
if __name__ == "__main__":
    # Create some accounts
    acc1 = Bank("Alice", 25, "Downtown", "F", 5000, "1001", 1234)
    acc2 = Bank("Bob", 30, "Uptown", "M", 3000, "1002", 5678)

    # Store all accounts in a list
    all_accounts = [acc1, acc2]

    # Test transfer
    acc1.display()
    acc1.transfer(all_accounts)  # Pass the list of accounts
    acc1.display()
    acc2.display()
```
